### Criando uma função com uma estratégia para oportunidades do mercado

```python
def oportunidade_media_setor(fiis, setor='Logística', label_setor='Setor'):

    media_setor = fiis_aux.groupby('Setor').agg(['mean','std'])

    fiis_setor = fiis[fiis[label_setor].isin([setor])]

    filter_ = \
            (fiis_setor['QuantidadeAtivos'] > 5) &\
            (fiis_setor['Liquidez Diária'] > 5000) &\
            (fiis_setor['P/VPA'] < 1.0) &\
            (fiis_setor['DY (12M)Acumulado'] > media_setor.loc[setor, ('DY (12M)Acumulado','mean')])

    print('média do setor Yield: {}'.format(media_setor.loc[setor, ('DY (12M)Acumulado','mean')]))
    print('média do setor p/VPA: {}'.format(media_setor.loc[setor, ('P/VPA','mean')]))
    print('média do setor Ativos: {}'.format(media_setor.loc[setor, ('QuantidadeAtivos','mean')]))

    return fiis_setor[filter_]
```

```python
oportunidade = oportunidade_media_setor(fiis_aux, setor='Logística')
oportunidade.sort_values('DY (12M)Acumulado', ascending=False, inplace=True)
oportunidade
```
